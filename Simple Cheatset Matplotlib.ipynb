# Matplotlib and Seaborn Cheatset

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

## Multiple Figure
### Plotting Multiple figure using Matplotlib Subplots

```python
x = np.random.randint(0,10,5)
y = np.random.randint(0,10,5)

fig, ax = plt.subplots(1, 2)
fig2,ax2 = plt.subplots(2, 2)
fig3, ax3 = plt.subplots(2)

#Vertically Stack
ax[0].plot(x,y)
ax[0].set_title('left')
ax[1].plot(x,-y,'green')
ax[1].set_title('right')

#Horizontally Stack
ax3[0].plot(x,y)
ax3[0].set_title('left')
ax3[1].plot(x,-y,'green')
ax3[1].set_title('right')

#Multiple
ax2[0,0].plot(x,y,'green')
ax2[0,0].set_title('topleft')
ax2[0,1].plot(x,y,'red')
ax2[0,1].set_title('topright')
ax2[1,0].plot(x,y,'blue')
ax2[1,0].set_title('botleft')
ax2[1,1].plot(x,y,'black')
ax2[1,1].set_title('botright')
```

# Jointplot
### Joint plot is both indicate scatter as well as histogram of the data representation the scatterplot represent joint distribution/correlation between 2 variables the histogram/distplot represent a marginal distribution which measure only single variable

```python
x = [1,2,3,4,5]
y = [6,7,8,9,10]
df = pd.DataFrame(list(zip(x, y)), columns =['x', 'y']) 
df.head()
```

```python
sns.jointplot(x='x',y='y',data=df)
```

## Dist Plot / Histogram
### Distplot is a way to approximate distribution of a single data / marginal distribution data

```python
sns.distplot(x=df['x'],kde=True,bins=5)
```

## KDE Plot
### KDE or kernel desnsity estimation plot  is a non-parametric way to estimate the probability density function of a random variable. Kernel density estimation is a fundamental data smoothing problem where inferences about the population are made, based on a finite data sample

```python
sns.kdeplot(data=df,x='x')
```

## Scatter Plot
### Scatter plot is a way to display values for typically two variables for a set of data (Joint Distribution)

```python
sns.scatterplot(x='x',y='y',data=df)
```

## Box Plot
### Boxplot is a graph that gives you a good indication of how the values in the data are spread out.

```python
df2 = sns.load_dataset('tips')
df2.head()
```

```python
plt = sns.boxplot(x = 'day', y ='tip',data=df2)
```

```python
dataX = [3,12,15,16,16,17,19,34]
dataY = ['data','data','data','data','data','data','data','data','data']
df2 = pd.DataFrame(list(zip(dataX, dataY)), columns =['x', 'y']) 
df2.head()
```

## Outlier of Boxplot
### Outlier indicate data is outside of the whisker using 1.5(IQR) Rules
### Dataframe 2 Value Min = 3, Max = 34, Q1 = 13.5, Q2/Median = 16, Q3 = 18
### Outer Outliers Q1 - 1.5(IQR) = 13.5 - 1.5(18-13.5) = 13.5 - 6.75 = 6.75
### Upper Outliers Q3 + 1.5(IQR) = 18 + 1.5(18-13.5) = 18 + 6.75 = 24.75, threfore data which less than outter outlier marked as outlier and data more than upper outlier marked as upper outlier, in this case Minimum is less than 6.75 therefore 3 is outter outlier, and 34 is more than 24.75 therefore 34 is outlier, and new minimum is 12 and new maximum is 19

```python
plt2 = sns.boxplot(x='x',y='y',data=df2)
```
